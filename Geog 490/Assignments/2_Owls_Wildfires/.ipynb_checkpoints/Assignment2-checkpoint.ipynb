### e. Reproject the dataset to UTM Zone 10N (EPSG:32610).
### Make a new column called pairs which has a value of 1 if there are both male and female owls in that location.


```python
SpottedOwls_utm = SpottedOwls.to_crs(epsg=32610)

ColumnNames = ['pairs']
pairsofowls = pd.DataFrame(0, index=np.arange(len(SpottedOwls)), columns=ColumnNames)


# Check if 'Primary' column already exists in file DataFrame
if 'pairs' not in SpottedOwls.columns:
    # Loop through each row in the file DataFrame
    for index, row in SpottedOwls.iterrows():
        # Check if the river is a primary river
        if row['males'] == 1.0 and row['females'] == 1.0:
            # If it is, set the value to 1 in the PrimeOrTributary DataFrame
            pairsofowls.loc[index, 'pairs'] = 1
        else:
            # Otherwise, set the value to 0
            pairsofowls.loc[index, 'pairs'] = 0

    # Concatenate the file DataFrame with the PrimeOrTributary DataFrame
    SpottedOwls = pd.concat([SpottedOwls, pairsofowls], axis=1)

# Print the updated DataFrame to verify the result
print(SpottedOwls)
```

```python
wildfire = gpd.read_file("C:/Users/jettr/Dropbox (University of Oregon)/23-24/Spring/Geog 490/Assignment 2/data/oregon_fires.shp")

SpottedOwls = gpd.read_file("C:/Users/jettr/Dropbox (University of Oregon)/23-24/Spring/Geog 490/Assignment 2/data/spotted_owls.shp")

wildfire_proj = wildfire.to_crs(epsg=32610)
spottedowls_proj = SpottedOwls.to_crs(epsg=32610) # they have to both be in the same projection
```

### a. How many owl pairs were affected by the wildfires?

```python
# Merge the two GeoDataFrames based on a common attribute (e.g., 'name' or 'ID')
merged_data = gpd.sjoin(wildfire_proj, spottedowls_proj, how='left', predicate='intersects') # merge datasets
# how = 'left' means that all data from the left will remain, while data from the right will stay if they intersect rows from the left, and are dropped if not
# data from the left may also be duplicated if there are multiple hits from the right dataset

# Drop rows with missing values (NaN)
merged_data = merged_data.dropna()
```

```python
merged_data.head()
```

```python
PairsOwlsGone = len(merged_data[ (merged_data['males'] == 0.0) | (merged_data['females'] == 0.0) ])

print('the amount of affected pairs of owls: ' + str(PairsOwlsGone))
```
